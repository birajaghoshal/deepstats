```python
mlp_fat = eval(str(np.load('../../Desktop/history_fat_mlp.npy')))
```

```python
def plot_history(ds, names):
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.title('Loss value')
    for d, name in zip(ds, names):
        plt.plot(d['loss'], label=name + ' | train_loss')
        plt.plot(d['val_loss'], label=name + ' | val_loss')
    plt.legend()
    plt.subplot(1,2,2)
    plt.title('Accuracy')
    for d, name in zip(ds, names):
        plt.plot(d['acc'], label=name + ' | train_acc')
        plt.plot(d['val_acc'], label=name + ' | val_acc')
    plt.legend(loc=4)
```

```python
plot_history([mlp, mlp_deep, pb, pbd], ['mlp', 'mlp_deep', 'bayes', 'drop_bayes'])
```

```python
plot_history([mlp_fat], ['mlp_fat'])
```
