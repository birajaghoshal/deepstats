```python
import pandas as pd
from collections import OrderedDict
```

```python
df_d = pd.read_csv('/Users/roliveira/Repos/deepstats/uncertainty_experiments.csv', sep=';')
df_b = pd.read_csv('/Users/roliveira/Repos/deepstats/bayesian_uncertainty_experiments.csv')
df_pb = pd.read_csv('/Users/roliveira/Repos/deepstats/poor_bayesian_uncertainty_experiments.csv')
df_bayes_mnist = pd.read_csv('/Users/roliveira/Repos/deepstats/bayesian_uncertainty_experiments_mnist.csv')
```

```python
cols = ['test_acc', 'Bayesian entropy bal_acc', 'Bayesian prediction STD bal_acc', 'Classical entropy bal_acc']
dropout = df_d.groupby(['dataset', 'inside_labels']).mean()[cols]
```

```python
df_b.loc[df_b.dataset == 'cifar10','dataset'] = 'cifar'
cols = ['train_time', 'test_acc', 'Bayesian entropy bal_acc', 'Bayesian prediction STD bal_acc']
bayesian = df_b.groupby(['dataset', 'inside_labels']).mean()[cols]
```

```python
df_bayes_mnist.loc[df_bayes_mnist.dataset == 'cifar10','dataset'] = 'cifar'
cols = ['train_time', 'test_acc', 'Bayesian entropy bal_acc', 'Bayesian prediction STD bal_acc']
bayesian_mnist_mean = df_bayes_mnist.groupby(['dataset', 'inside_labels', 'unknown_labels']).mean()[cols]
bayesian_mnist_std = df_bayes_mnist.groupby(['dataset', 'inside_labels', 'unknown_labels']).std()[cols]
```

```python
df_pb.loc[df_pb.dataset == 'cifar10','dataset'] = 'cifar'
cols = ['train_time', 'test_acc', 'Bayesian entropy bal_acc', 'Bayesian prediction STD bal_acc']
poor_bayesian = df_pb.groupby(['dataset', 'inside_labels']).mean()[cols]
```

```python
cols = ['dataset','labels','dropout_test_acc','bayesian_test_acc','dropout_entropy','dropout_std',
        'dropout_entropy_classical','bayesian_entropy','bayesian_std']
results = pd.DataFrame(columns=cols)
for ((drop_key, *drop_row), (bayes_key, *bayes_row)) in zip(dropout.itertuples(), bayesian.itertuples()):
    assert drop_key == bayes_key
    row = OrderedDict()
    row['dataset'] = [drop_key[0]]
    row['labels'] = [drop_key[1]]
    row['dropout_test_acc'] = [drop_row[0]]
    row['bayesian_test_acc'] = [bayes_row[0]]
    
    row['dropout_entropy'] = [drop_row[1]]
    row['dropout_std'] = [drop_row[2]]
    row['dropout_entropy_classical'] = [drop_row[3]]
    
    row['bayesian_entropy'] = [drop_row[1]]
    row['bayesian_std'] = [drop_row[2]]
    
    values = [drop_key[0], drop_key[1], drop_row[0], bayes_row[0], drop_row[1], drop_row[2], drop_row[3],
              bayes_row[1], bayes_row[2]] 
    
    results.loc[len(results)] = values
```

```python
cols = ['dataset','labels', 'poor_bayesian_time', 'bayesian_time', 'poor_bayesian_test_acc', 'bayesian_test_acc', 
        'bayesian_entropy','bayesian_std', 'poor_bayesian_entropy','poor_bayesian_std']
results = pd.DataFrame(columns=cols)
for ((poor_key, *poor_row), (bayes_key, *bayes_row)) in zip(poor_bayesian.itertuples(), bayesian.itertuples()):
    assert poor_key == bayes_key    
    values = [poor_key[0], poor_key[1], 
              poor_row[0], bayes_row[0],
              poor_row[1], bayes_row[1],
              poor_row[2], bayes_row[2],
              poor_row[3], bayes_row[3]] 
    
    results.loc[len(results)] = values
```

```python
cols = ['dataset','inside_labels', 'unknown_labels',
        #'train_time', 'train_time_std',
        'test_acc', 'test_acc_std', 
        'bayes_entropy','bayes_entropy_std', 
        'bayesian_std', 'bayes_std_std']
results = pd.DataFrame(columns=cols)
for ((mean_key, *mean_row), (std_key, *std_row)) in zip(bayesian_mnist_mean.itertuples(), 
                                                        bayesian_mnist_std.itertuples()):
    assert mean_key == std_key    
    values = [mean_key[0], mean_key[1], mean_key[2],
              #mean_row[0], std_row[0],
              mean_row[1], std_row[1],
              mean_row[2], std_row[2],
              mean_row[3], std_row[3]] 
    
    results.loc[len(results)] = values
```
